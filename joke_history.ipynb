```python
import os
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()

client = OpenAI(
    base_url="https://api.groq.com/openai/v1",
    api_key=os.environ.get("GROQ_API_KEY")
)

MODEL="llama-3.1-8b-instant"
```

```python

# Add history what you usually put as messages
history = [
    {"role": "user", "content": "Tell me a knock-knock joke."}
]

# Put reponse and add history as the message
response = client.chat.completions.create(
    model=MODEL,
    messages=history
)

# Generate the reply and print it in a format
assistant_reply = response.choices[0].message.content
print("AI:", assistant_reply)

# Add previous response and follow up question to the history
history.append({"role": "assistant", "content": assistant_reply})
history.append({"role": "user", "content": "Who is there?"})

# Now put in the second input and generate response
second_response = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=history
)

print("AI Follow-up:", second_response.choices[0].message.content)
```
