```python

def count_tokens(text: str, encoding_name: str = "cl100k_base") -> int:
    """Returns the number of tokens in a text string using tiktoken."""
    encoding = tiktoken.get_encoding(encoding_name)
    num_tokens = len(encoding.encode(text))
    return num_tokens

def main():
    # define the model and prompt
    model_name = "llama-3.1-8b-instant" 
    prompt = "Explain the importance of tokenization in large language models in two sentences."

    # before sending req, estimate no of tokens
    estimated_tokens = count_tokens(prompt)
    print(f"--- Token Count Analysis ---")
    print(f"Estimated prompt tokens (via tiktoken): {estimated_tokens}\n")

    print(f"--- Sending Request to Groq ({model_name}) ---")
    # standard chat completion request
    chat_completion = client.chat.completions.create(
        messages=[
            {
                "role": "user",
                "content": prompt,
            }
        ],
        model=model_name,
    )

    # response from the model
    response_text = chat_completion.choices[0].message.content
    print(f"Response:\n{response_text}\n")

    # exact tokens returned by groq
    print(f"--- Exact API Usage Stats ---")
    print(f"Prompt Tokens used: {chat_completion.usage.prompt_tokens}")
    print(f"Completion Tokens used: {chat_completion.usage.completion_tokens}")
    print(f"Total Tokens used: {chat_completion.usage.total_tokens}")

if __name__ == "__main__":
    main()
```
