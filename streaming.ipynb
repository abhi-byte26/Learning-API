```python
import os
import sys
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()

client = OpenAI(
    base_url="https://api.groq.com/openai/v1",
    api_key=os.environ.get("GROQ_API_KEY")
)

MODEL="llama-3.1-8b-instant"
```

```python

def main():
    # a repetitive task for demonstration
    prompt = "Recite the tongue twister 'Peter Piper picked a peck of pickled peppers' five times."
    
    print(f"--- Sending Prompt to Groq ({MODEL_NAME}) ---")
    print(f"Prompt: \"{prompt}\"\n")
    print("--- Streamed Response Start ---")

    # completion is triggered with stream=true
    stream = client.chat.completions.create(
        model=MODEL_NAME,
        messages=[
            {
                "role": "user",
                "content": prompt,
            }
        ],
        stream=True,  # the trigger
    )

    # iterate stream events as they arrive
    for chunk in stream:
        # check if current chunk has content to print
        if chunk.choices[0].delta.content is not None:
            # print each chunk on arrival
            # end="" makes sure no newline is added in bw tokens
            # flush=True forces the console to display the output immediately
            print(chunk.choices[0].delta.content, end="", flush=True)

    print("\n\n--- Streamed Response End ---")

if __name__ == "__main__":
    main()
```
